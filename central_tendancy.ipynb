```python
# create a historgram of fixed acidity feature
plt.figure(figsize=(11,6))
sns.histplot(data = data, x = 'fixed acidity', color = 'orange', edgecolor = 'linen', alpha = 0.5, bins = 20)
plt.title('Fixed Acidity Distribution', fontsize = 20)
plt.xlabel('Fixed Acidity', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
#histogram is normally distributed. 
#now lets check the central tendancy of the fixed acidity feature.
```

```python
round(data['fixed acidity'].mean(), 2)
```

```python
data['fixed acidity'].median()
```

```python
plt.figure(figsize=(11,6))
sns.histplot(data = data, x = 'fixed acidity', color = 'orange', edgecolor = 'linen', alpha = 0.5, bins = 20)
plt.title('Fixed Acidity Distribution', fontsize = 20)
plt.xlabel('Fixed Acidity', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.vlines(data['fixed acidity'].mean(), ymin =0, ymax = 2000, color = 'red', label = 'Mean', linestyle = '--')
plt.vlines(data['fixed acidity'].median(), ymin =0, ymax = 2000, color = 'blue', label = 'Median', linestyle = '--')
plt.legend()
plt.show()
```

```python
# volatile acidity feature
plt.figure(figsize=(11,6))
sns.histplot(data=data, x = 'volatile acidity', color = 'green', edgecolor = 'linen', alpha = 0.5, bins = 20)
plt.title('Volatile Acidity Distribution', fontsize = 20)
plt.xlabel('Volatile Acidity', fontsize = 15)
plt.ylabel('Count', fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(11,6))
sns.distplot(data['volatile acidity'], color = 'green')
plt.title('Volatile Acidity Distribution', fontsize = 20)
plt.xlabel('Volatile Acidity', fontsize = 15)
plt.ylabel('Density', fontsize = 15)
plt.show()
```
