```python
plt.figure(figsize=(10,6))
sns.scatterplot(data= data, x = "age", y = "salary", color = "blue", alpha = 0.7)
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = data, x  = "healthy_eating", y = "active_lifestyle", color = "green", alpha = 0.7)
plt.title("Healthy Eating vs Active Lifestyle")
plt.xlabel("Healthy Eating")
plt.ylabel("Active Lifestyle")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = data, x = "healthy_eating", y = 'salary', color = "orange", alpha = 0.7)
plt.title("Healthy Eating vs Salary")
plt.xlabel("Healthy Eating")    
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data = data, x = "groups", palette="Set2")
plt.title("Distribution of Groups")
plt.xlabel("Groups")
plt.ylabel("Count")
plt.show()
```

```python
sns.histplot(data['salary'], bins=15, kde=True, palette="Set1")
plt.title("Distribution of Salary")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```
